```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


imagen = cv2.imread('taj.jpg', cv2.IMREAD_GRAYSCALE)


laplaciano = cv2.Laplacian(imagen, cv2.CV_64F)
laplaciano = np.uint8(np.absolute(laplaciano))


gauss = cv2.GaussianBlur(imagen, (1, 1), 4)


log = cv2.Laplacian(gauss, cv2.CV_64F)
log = np.uint8(np.absolute(log))


plt.figure(figsize=(12, 4))


plt.subplot(1, 3, 1)
plt.title("Original (Grises)")
plt.imshow(imagen, cmap='gray')
plt.axis('off')


plt.subplot(1, 3, 2)
plt.title("Laplaciano")
plt.imshow(laplaciano, cmap='gray')
plt.axis('off')


plt.subplot(1, 3, 3)
plt.title("LoG (Laplaciano del Gaussiano)")
plt.imshow(log, cmap='gray')
plt.axis('off')


plt.tight_layout()
plt.show()
```
