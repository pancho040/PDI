```python
def calcular_factorial():
    n = int(input("Ingrese un número: "))
    resultado = 1 if n == 0 else 1
    for i in range(1, n + 1):
        resultado *= i
    print(f"El factorial de {n} es {resultado}")

def es_primo():
    n = int(input("Ingrese un número: "))
    if n < 2:
        print(f"{n} no es primo")
        return
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            print(f"{n} no es primo")
            return
    print(f"{n} es primo")

def encontrar_mayor():
    numeros = [int(input(f"Ingrese el número {i + 1}/3: ")) for i in range(3)]
    print(f"El número mayor ingresado es {max(numeros)}")

def es_palindromo():
    cadena = input("Ingrese una cadena: ")
    if cadena == cadena[::-1]:
        print(f"{cadena} es un palíndromo.")
    else:
        print(f"{cadena} no es un palíndromo.")

def contar_repeticiones():
    numeros = [int(input(f"Ingrese un número {i + 1}/7: ")) for i in range(7)]
    num_mas_repetido = max(set(numeros), key=numeros.count)
    repeticiones = numeros.count(num_mas_repetido)
    print(f"El número que más se repite es {num_mas_repetido}, con {repeticiones} repeticiones.")

def menu():
    opciones = {
        1: calcular_factorial,
        2: es_primo,
        3: encontrar_mayor,
        4: es_palindromo,
        5: contar_repeticiones
    }

    while True:
        print("\nSeleccione una opción:")
        print("1. Calcular Factorial")
        print("2. Verificar si es Primo")
        print("3. Encontrar el Mayor")
        print("4. Verificar si es Palíndromo")
        print("5. Contar Repeticiones")
        print("6. Salir")

        try:
            opcion = int(input("Ingrese el número de la opción: "))
            if opcion == 6:
                print("Fin del programa.")
                break
            elif opcion in opciones:
                opciones[opcion]()
            else:
                print("Opción no válida, por favor intente nuevamente.")
        except ValueError:
            print("Por favor, ingrese un número válido.")

menu()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def convertir_rgb(img, output_path="imagen_rgb.jpg"):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    cv2.imwrite(output_path, img_rgb)
    plt.figure(figsize=(5, 5))
    plt.imshow(img_rgb)
    plt.title('BGR a RGB')
    plt.axis('off')
    plt.show()

def escala_grises(img, output_path="imagen_gris.jpg"):
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imwrite(output_path, img_gray)
    plt.figure(figsize=(5, 5))
    plt.imshow(img_gray, cmap="gray")
    plt.title('Escala de Grises')
    plt.axis('off')
    plt.show()

def mostrar_canales_gris(img):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    canales = ['Rojo', 'Verde', 'Azul']
    
    plt.figure(figsize=(10, 8))
    plt.subplot(2, 2, 1)
    plt.imshow(img_rgb)
    plt.title('Imagen Original')
    plt.axis('off')

    for i in range(3):
        plt.subplot(2, 2, i + 2)
        plt.imshow(img[:, :, i], cmap='gray')
        plt.title(f'Canal {canales[i]}')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

def mostrar_canales_rgb(img):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    r, g, b = img[:, :, 2], img[:, :, 1], img[:, :, 0]

    r_img, g_img, b_img = np.zeros_like(img_rgb), np.zeros_like(img_rgb), np.zeros_like(img_rgb)
    r_img[:, :, 0], g_img[:, :, 1], b_img[:, :, 2] = r, g, b

    plt.figure(figsize=(10, 8))
    plt.subplot(2, 2, 1)
    plt.imshow(img_rgb)
    plt.title('Original')
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.imshow(r_img)
    plt.title('Canal Rojo')
    plt.axis('off')

    plt.subplot(2, 2, 3)
    plt.imshow(g_img)
    plt.title('Canal Verde')
    plt.axis('off')

    plt.subplot(2, 2, 4)
    plt.imshow(b_img)
    plt.title('Canal Azul')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

def mostrar_canales_colormap(img):
    b, g, r = cv2.split(img)

    plt.figure(figsize=(10, 8))
    plt.subplot(2, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Original')
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.imshow(r, cmap='Reds')
    plt.title('Canal Rojo')
    plt.axis('off')

    plt.subplot(2, 2, 3)
    plt.imshow(g, cmap='Greens')
    plt.title('Canal Verde')
    plt.axis('off')

    plt.subplot(2, 2, 4)
    plt.imshow(b, cmap='Blues')
    plt.title('Canal Azul')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

img = cv2.imread('perrito.jpg')

convertir_rgb(img)
escala_grises(img)
mostrar_canales_gris(img)
mostrar_canales_rgb(img)
mostrar_canales_colormap(img)
```

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

fig, axs = plt.subplots(2, 5, figsize=(30, 10))

img_bgr = cv2.imread('1.jpg', cv2.IMREAD_COLOR)

bayer_filters = [
    cv2.COLOR_BAYER_BG2RGB,
    cv2.COLOR_BAYER_GB2RGB,
    cv2.COLOR_BAYER_RG2RGB,
    cv2.COLOR_BAYER_GR2RGB,
    cv2.COLOR_BAYER_BG2BGR,
    cv2.COLOR_BAYER_GB2BGR,
    cv2.COLOR_BAYER_RG2BGR,
    cv2.COLOR_BAYER_GR2BGR
]

titles = [
    'Original', 'Bayer BG to RGB', 'Bayer GB to RGB', 'Bayer RG to RGB', 'Bayer GR to RGB',
    'Bayer BG to BGR', 'Bayer GB to BGR', 'Bayer RG to BGR', 'Bayer GR to BGR'
]

axs[0, 0].imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title('Original')
axs[0, 0].axis('off')

for ax, bayer_filter, title in zip(axs.flat[1:], bayer_filters, titles[1:]):
    try:
        img = cv2.cvtColor(img_bgr, bayer_filter)
        ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    except cv2.error:
        ax.imshow(img_bgr, cmap='gray')
    ax.set_title(title)
    ax.axis('off')

plt.show()
```
