```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def convolucion_media_horizontal(imagen,tamanioKernel):
    kernel = np.ones((1,tamanioKernel),np.float32)/tamanioKernel
    imgfin = cv2.filter2D(imagen,-1,kernel)
    return imgfin

imagen = cv2.imread('distorsionada.jpeg')
gris = cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)

img_rest = convolucion_media_horizontal(gris,25)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB))

plt.subplot(1,2,2)
plt.imshow(img_rest,cmap='gray')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def convolucion_media_horizontal(imagen,tamanioKernel):
    kernel = np.ones((1,tamanioKernel),np.float32)/tamanioKernel
    imgfin = cv2.filter2D(imagen,-1,kernel)
    return imgfin

imagen = cv2.imread('distorsionada.jpeg')
img_rgb= cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)

imgrest_r = convolucion_media_horizontal(img_rgb[:,:,0],25)
imgrest_g = convolucion_media_horizontal(img_rgb[:,:,1],25)
imgrest_b = convolucion_media_horizontal(img_rgb[:,:,2],25)

imgcolor = np.stack([imgrest_r,imgrest_g,imgrest_b],axis=2)


plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img_rgb)

plt.subplot(1,2,2)
plt.imshow(imgcolor)

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_ruido = np.random.randint(0,256,size=[256,256],dtype=np.uint8)

img_blur = cv2.GaussianBlur(img_ruido,(5,5),0)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img_ruido,cmap='gray')

plt.subplot(1,2,2)
plt.imshow(img_blur,cmap='gray')

plt.show()
```
