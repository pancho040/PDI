```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize']=[10,5]
img=cv2.imread('perrito.jpg',cv2.IMREAD_COLOR)
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
r=img[:,:,2]
g=img[:,:,1]
b=img[:,:,0]
r_imgen=np.zeros_like(img_rgb)
r_imgen[:,:,0]=r
g_imgen=np.zeros_like(img_rgb)
g_imgen[:,:,1]=g
b_imgen=np.zeros_like(img_rgb)
b_imgen[:,:,2]=b

plt.subplot(141)
plt.title('Original')
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(142)
plt.title('CAnal rojo')
plt.imshow(r_imgen)
plt.axis('off')

plt.subplot(143)
plt.title('CAnal Verde')
plt.imshow(g_imgen)
plt.axis('off')

plt.subplot(144)
plt.title('CAnal Azul')
plt.imshow(b_imgen)
plt.axis('off')

plt.show()
```

```python
import cv2

# Obtener todos los atributos de cv2 que contienen "BAYER"
bayer_filters = [f for f in dir(cv2) if "BAYER" in f]

# Mostrar la lista de filtros
print(bayer_filters)
```

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

fig, axs = plt.subplots(2, 4, figsize=(25, 10))
img_bgr = cv2.imread('1.jpg', cv2.IMREAD_GRAYSCALE)

bayer_filters = [
    cv2.COLOR_BAYER_BG2RGB,
    cv2.COLOR_BAYER_GB2RGB,
    cv2.COLOR_BAYER_RG2RGB,
    cv2.COLOR_BAYER_GR2RGB,
    cv2.COLOR_BAYER_BG2BGR,
    cv2.COLOR_BAYER_GB2BGR,
    cv2.COLOR_BAYER_RG2BGR,
    cv2.COLOR_BAYER_GR2BGR
]

titles = [
    'Bayer BG to RGB', 'Bayer GB to RGB', 'Bayer RG to RGB', 'Bayer GR to RGB',
    'Bayer BG to BGR', 'Bayer GB to BGR', 'Bayer RG to BGR', 'Bayer GR to BGR'
]

for ax, bayer_filter, title in zip(axs.flat, bayer_filters, titles):
    img = cv2.cvtColor(img_bgr, bayer_filter)
    ax.imshow(img)
    ax.set_title(title)
    ax.axis('off')

plt.show()
```
