```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('finn.jpg', cv2.IMREAD_GRAYSCALE)


lap_kernel = np.array([
    [0,  1, 0],
    [1, -4, 1],
    [0,  1, 0]
], dtype=np.float32)

lap = cv2.filter2D(img, cv2.CV_64F, lap_kernel)
lap = np.uint8(np.absolute(lap))


blur = cv2.GaussianBlur(img, (5, 5), 0)
log = cv2.filter2D(blur, cv2.CV_64F, lap_kernel)
log = np.uint8(np.absolute(log))

plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Laplaciano")
plt.imshow(lap, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("LoG")
plt.imshow(log, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
